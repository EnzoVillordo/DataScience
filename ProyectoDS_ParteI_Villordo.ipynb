```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/EnzoVillordo/DataScience/main/employee_survey.csv"
df = pd.read_csv(url)

# para verificar
df.head()
```

1- Analisis univariado

```python
# Calcular las frecuencias
gender_counts = df['Gender'].value_counts()

# Crear grÃ¡fico de torta
plt.figure(figsize=(6,6))
plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct='%1.1f%%',      # muestra el porcentaje con un decimal
    startangle=90,          # inicia el primer sector en 90 grados
    colors=['blue','lightpink', 'lightblue']  # opcional: colores personalizados
)
plt.title('DistribuciÃ³n de empleados por gÃ©nero')
plt.show()
```

```python
sns.countplot(x='JobLevel', data=df)
plt.title('Nivel del puesto')
ax = sns.countplot(x='JobLevel', data=df)
ax.bar_label(ax.containers[0], padding=3)
plt.show()
```

2- 1er Analisis Bivariado

```python
sns.countplot(data=df, x="Dept", hue="Gender")
plt.xticks(rotation=45); plt.show()
```

2do Analisis Bivariado : 2 Ejemplos de lo mismo

```python
sns.boxplot(data=df, x="JobLevel", y="JobSatisfaction")
plt.xlabel("Nivel del puesto")
plt.ylabel("SatisfacciÃ³n laboral")
plt.title("RelaciÃ³n entre Nivel del puesto y SatisfacciÃ³n laboral")
plt.show()
```

```python
sns.violinplot(data=df, x="JobLevel", y="JobSatisfaction")
plt.xlabel("Nivel del puesto")
plt.ylabel("SatisfacciÃ³n laboral")
plt.title("DistribuciÃ³n de la SatisfacciÃ³n laboral por Nivel del puesto")
plt.show()
```

3- 1er Analisis: Multivariado

```python
sns.scatterplot(
    data=df,
    x="SleepHours",
    y="PhysicalActivityHours",
    hue="Stress",       # Color segÃºn nivel de estrÃ©s
    palette="coolwarm",
    size="Stress",      # TamaÃ±o de los puntos segÃºn estrÃ©s
    alpha=0.7
)
plt.xlabel("Horas de sueÃ±o")
plt.ylabel("Horas de actividad fÃ­sica")
plt.title("RelaciÃ³n entre SueÃ±o, Actividad FÃ­sica y EstrÃ©s")
plt.show()
```
